***🔍 Face Analysis Using DeepFace (Age, Gender, Emotion Detection)***

```python
# 📦 Step 1: Install Required Libraries
!pip install deepface
!pip install opencv-python-headless

# 🧠 Step 2: Import Libraries
from google.colab import files
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# 📤 Step 3: Upload Image
print("📤 Please upload an image of a face...")
uploaded = files.upload()

# 🖼️ Step 4: Read and Display Image
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title("📸 Uploaded Image")
plt.show()

# 🔍 Step 5: Analyze Image using DeepFace
print("🔍 Analyzing image. Please wait...")
results = DeepFace.analyze(img_path=img_path, actions=['age', 'gender', 'emotion'], enforce_detection=True)

# 📊 Step 6: Extract Predictions
predicted_age = results[0]['age']
predicted_gender = results[0]['dominant_gender']
predicted_emotion = results[0]['dominant_emotion']

# 📝 Step 7: Display Results
print(f"👶 Predicted Age: {predicted_age}")
print(f"🧑 Predicted Gender: {predicted_gender}")
print(f"😊 Predicted Emotion: {predicted_emotion}")
```
