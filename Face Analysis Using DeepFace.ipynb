***ðŸ” Face Analysis Using DeepFace (Age, Gender, Emotion Detection)***

```python
# ðŸ“¦ Step 1: Install Required Libraries
!pip install deepface
!pip install opencv-python-headless

# ðŸ§  Step 2: Import Libraries
from google.colab import files
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# ðŸ“¤ Step 3: Upload Image
print("ðŸ“¤ Please upload an image of a face...")
uploaded = files.upload()

# ðŸ–¼ï¸ Step 4: Read and Display Image
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title("ðŸ“¸ Uploaded Image")
plt.show()

# ðŸ” Step 5: Analyze Image using DeepFace
print("ðŸ” Analyzing image. Please wait...")
results = DeepFace.analyze(img_path=img_path, actions=['age', 'gender', 'emotion'], enforce_detection=True)

# ðŸ“Š Step 6: Extract Predictions
predicted_age = results[0]['age']
predicted_gender = results[0]['dominant_gender']
predicted_emotion = results[0]['dominant_emotion']

# ðŸ“ Step 7: Display Results
print(f"ðŸ‘¶ Predicted Age: {predicted_age}")
print(f"ðŸ§‘ Predicted Gender: {predicted_gender}")
print(f"ðŸ˜Š Predicted Emotion: {predicted_emotion}")
```
